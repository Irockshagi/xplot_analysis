### Introduction

This is exercise is an example for viewing logs using iPython Notebook. The exercise is divided into several parts: loading the function for viewing the logs, loading the LAS file, calculating some derivative logs, and generate the logview. The first three items are explained in detailed at different article on "Cross plot analysis using iPython Notebook and will not be repeated here. The last item, generating the logview, will be explained later.

Let us start by loading the logview function. The input parameters are:

- dep : can be measured depth (MD), true vertical depth (TVD), or true vertical depth subsea (TVDSS)
- dep1 : the depth of top of interval we want to view
- dep2 : the depth of base of interval we want to view
- dinp : matrix of logs to be viewed
- dmark : list of well markers in the related depth (MD, TVD, or TVDSS)
- dlab : labels of logs, must be the same with the number of logs

```python
def mylogview(dep,dep1,dep2,dinp,dmark,dlab):
    import numpy as np
    import matplotlib.pyplot as plt
    %matplotlib inline
    nlog,ndat = np.shape(dinp)
    nmark = len(dmark)
    sizey = 8
    sizex = nlog*2
    fig = plt.figure()
    fig.set_size_inches(sizex,sizey)
    for i in range(0,nlog):
        xm = [np.min(dinp[i][:]), np.max(dinp[i][:])]
        plt.subplot(1,nlog,i+1)
        plt.plot(dinp[i][:],dep,'b')
        for k in range(0,nmark):
            ym = [dmark[k],dmark[k]]
            plt.plot(xm,ym,'r')
        plt.xlabel(dlab[i])
        if (i==0):
            plt.ylabel('Depth (m)')
        plt.ylim(dep1,dep2)
        plt.gca().invert_yaxis()
    plt.show()
    return
```

After loading the logview function, we load the logs exist in the LAS file. The names of the logs written below are similar with the curve information described in the LAS file (see for example "Cross plot analysis using iPython Notebook").

```python
import numpy as np
from las import LASReader

inpdat = LASReader('well_2.las',null_subs=np.nan)

depmin = inpdat.start
depmax = inpdat.stop

dep = inpdat.data['DEPT']
vp = inpdat.data['Vp']
vs = inpdat.data['Vs']
gr = inpdat.data['GR']
rho = inpdat.data['RHOB']
nphi = inpdat.data['NPHI']
```

We can calculate some derivative logs for example P-impedance (ai), S-impedance (si), LambdaRho (lrho), MuRho (mrho), Vp/Vs (vpvs), and porosity from density (pord)

```python
ai = rho*vp*1000
si = rho*vs*1000
lrho = ai**2 - 2*si**2
mrho = si**2
vpvs = vp/vs
pord = (2.65-rho)/(2.65-1.05)
```

Suppose we want to view gamma ray and density logs. To do this, we generate matrix of logs (dinp), the labels (dlab), and the known well markers are stored in dmark. We want to display the whole interval (between depmin and depmax), and show the markers. To do this, execute mylogview using the parameters below

```python
dinp = [gr,rho]
dlab = ['GR','RHO']
dmark = [2153, 2183]
mylogview(dep,depmin,depmax,dinp,dmark,dlab)
```

We are interested only to view interval of interest. To do this, let us pick the top of interval to be 2120 m and 2220 m. The logview can be done now by the command below.

```python
mylogview(dep,2120,2220,dinp,dmark,dlab)
```

If we want to view more logs, we can do it simply by adding the logs to the matrix of logs (dinp), and therefore adding the labels in dlab. For example, we want to add P-impedance, S-impedance, Vp/Vs, and porosity logs to view. It is simple! Just execute the command below.

```python
dinp = [gr,rho,ai,si,vpvs,pord]
dlab = ['GR','RHO','P-impedance','S-impedance','Vp/Vs','Porosity']
dmark = [2153, 2183]
mylogview(dep,2120,2220,dinp,dmark,dlab)
```

The code is not perfect and may not be what you want. To improve your logview, for example improving the scale in the horizontal axis, you can learn about the matplotlib and use proper commands to make a better logview. However, it is shown here that you can do it easily using iPython Notebook.
